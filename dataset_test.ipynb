```python
import numpy as np
import os
import cv2
from os import listdir
import matplotlib.pyplot as plt

from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
from zipfile import ZipFile as zipFileMod

file_name = "datasetAnimals.zip"

# opening the zip file in READ mode
with zipFileMod(file_name, 'r') as zip_folder:
    # printing all the contents of the zip file
    zip_folder.printdir()

    # extracting all the files
    print('Extracting all the files now...')
    zip_folder.extractall()
    print('Done!')

num_classes = 10

# img_test = plt.imread("images/train/angry/0.jpg")
# plt.imshow(img_test)
```

```python
def labelNumber(label):
    match label:
        case "dog":
            return 0
        case "cat":
            return 1
        case "wild":
            return 2
```

```python
#Import the train dataset
num_classes = len(listdir("afhq/train"))
print("number of labels :", num_classes)
X_train = []
Labels_train = []
NamesCollections = []

for k,label in enumerate(listdir("afhq/train")):
  for image in listdir("afhq/train/" + label):
    img = cv2.imread("afhq/train/" + label + "/" + image, 0)
    res = cv2.resize(img, dsize=(28,28))
    X_train.append(res)
    labelNum = labelNumber(label)
    Labels_train.append(labelNum)
    NamesCollections.append("afhq/train/" + label + "/" + image)

X_train = np.array(X_train)
Labels_train = np.array(Labels_train)
print(X_train.shape)
print(Labels_train.shape)
```

```python
#Import the test dataset
num_classes = len(listdir("afhq/val"))
print("number of labels :", num_classes)
X_test = []
Labels_test = []
NamesCollections = []

for k,label in enumerate(listdir("afhq/val")):
  for image in listdir("afhq/val/" + label):
    img = cv2.imread("afhq/val/" + label + "/" + image, 0)
    res = cv2.resize(img, dsize=(28,28))
    X_test.append(res)
    labelNum = labelNumber(label)
    Labels_test.append(labelNum)
    NamesCollections.append("afhq/val/" + label + "/" + image)

X_test = np.array(X_test)
Labels_test = np.array(Labels_test)
print(X_test.shape)
print(Labels_test.shape)
```

```python
#change the shape of the object labels
y_train = to_categorical(Labels_train, num_classes) 
y_test = to_categorical(Labels_test, num_classes) 

# define the model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(num_classes, activation='softmax'))

# compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)

predictions = model.predict(X_test)

# calculate the accuracy of the predictions
accuracy = np.mean(np.argmax(predictions, axis=1) == np.argmax(y_test, axis=1))

print('Accuracy:', accuracy)
```
